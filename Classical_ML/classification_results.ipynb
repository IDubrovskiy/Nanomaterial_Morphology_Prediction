```python
import pandas as pd
import numpy as np
from optimize_classification import calculate_final_metrics, calculate_final_metrics_xgb
```

```python
particle_list = [
    "Cube",
    "Stick",
    "Sphere",
    "Flat",
    "Amorphous",
    "Cube_s",
    "Cube_m",
    "Cube_l",
    "Sphere_s",
    "Sphere_m",
    "Sphere_l",
    "Stick_s",
    "Stick_m",
    "Stick_l",
    "Flat_s",
    "Flat_m",
    "Flat_l",
    "Amorphous_s",
    "Amorphous_m",
    "Amorphous_l",
]
```

```python
final_df = pd.DataFrame(
        columns=[
            "rf_acc_unoptimized",
            "rf_f1_unoptimized",
            "rf_acc_optimized",
            "rf_f1_optimized",
        ])
for particle in particle_list:
    df = calculate_final_metrics(particle)
    final_df = pd.concat([final_df, df])
```

```python
final_df
```

```python
final_df['particle'] = particle_list
```

```python
final_df.to_csv('results_rf_filtered_15testsize.csv')
```

```python
final_df_xgb = pd.DataFrame(
        columns=[
            "xgb_acc_unoptimized",
            "xgb_f1_unoptimized",
            "xgb_acc_optimized",
            "xgb_f1_optimized",
        ])
for particle in particle_list:
    df = calculate_final_metrics_xgb(particle)
    final_df_xgb = pd.concat([final_df_xgb, df])
```

```python
final_df_xgb['particle'] = particle_list
```
